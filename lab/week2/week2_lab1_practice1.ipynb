```python
# calculate average
avg_rating_df = google_df[['App', 'Rating']].groupby(['App']).mean()
avg_rating_df = avg_rating_df.sort_values('Rating', ascending=False)
avg_rating_df.head(10)
```

```python
# calculate average
avg_rating_df1 = google_df1[['App', 'Rating']].groupby(['App']).mean()
avg_rating_df1 = avg_rating_df1.sort_values('Rating', ascending=False)
avg_rating_df1.head(10)
```

```python
# calculate average with count
avg_rating_df = google_df.groupby(['App'])['Rating'].agg(['mean', 'count'])
avg_rating_df.rename(columns={'mean': 'vote_average', 'count': 'vote_count'}, inplace=True)
avg_rating_df.sort_values('vote_average', ascending=False).head(10)
```

```python
# calculate average with count
avg_rating_df1 = google_df1.groupby(['App'])['Rating'].agg(['mean', 'count'])
avg_rating_df1.rename(columns={'mean': 'vote_average', 'count': 'vote_count'}, inplace=True)
avg_rating_df1.sort_values('vote_average', ascending=False).head(10)
```

```python
# weight function
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# weight function
def weighted_rating(x, m1=m1, C1=C1):
    v1 = x['vote_count']
    R1 = x['vote_average']
    return (v1/(v1+m1) * R1) + (m1/(m1+v1) * C1)
```

```python
# check the new scores
avg_rating_df2 = avg_rating_df.copy().loc[avg_rating_df['vote_count'] >= m]
avg_rating_df2['score'] = avg_rating_df2.apply(weighted_rating, axis=1)
avg_rating_df2.sort_values('score', ascending=False).head(10)
```
