```python
# import libraries
import pandas as pd

# load the datasets
# movie metadata
movies_meta = pd.read_csv('movies_metadata.csv', low_memory=False)
# and the keywords
movies_ratings = pd.read_csv('ratings.csv', low_memory=False)

# quick preview
movies_ratings.head()
```

```python
movies_meta.head()
```

```python
movies_meta = movies_meta.drop([19730, 29503, 35587])
movies_meta['id'] = movies_meta['id'].astype('int')
```

```python
# rename the column name in movies_meta

movies_meta.rename(columns = {'id':'movieId'}, inplace = True) 

movies_meta.head()
```

```python
# first need to combine the two 
movies_merged = movies_ratings.merge(movies_meta, on='movieId')
movies_merged.head()
```

```python
# see what a specific user e.g 110 rated
movies_merged[movies_merged['userId'] == 110].head(4)
```

```python
from itertools import permutations

# create a function that makes combinations of the movies that a user (x) reviewed
def create_combinations(x):
    combinations = pd.DataFrame(list(permutations(x.values, 2)), columns=['movie_a', 'movie_b'])
    return combinations
```

```python
# how many rating by user
avg_user = movies_merged.groupby(['userId']).size().reset_index(name='counts')
avg_user = avg_user.set_index('userId')
avg_user
```

```python
# more than 2 movies rated
avg_user = avg_user[avg_user['counts'] > 2]
users_index = avg_user.index

smaller_selection = movies_merged.iloc[users_index]
smaller_selection.head()
```

```python
smaller_selection.shape
```

```python
smaller_selection.columns
```

```python
# now using the function we created in the beginning
# use the create_combinations function

movie_combinations = smaller_selection.groupby('userId')['title'].apply(create_combinations)
movie_combinations.head(4)
```

```python
# calculate the occurences of the movie a and movie b

movie_combinations = movie_combinations.reset_index(drop=True)
combi_count = movie_combinations.groupby(['movie_a', 'movie_b']).size()
combi_count = combi_count.to_frame(name='size').reset_index()
combi_count.sort_values('size', ascending=False).head(5)
```

```python
# search for a specific movie to check 
combi_count[combi_count['movie_a'] == 'Three Colors: Red'].sort_values('size', ascending=False).head(10)
```
